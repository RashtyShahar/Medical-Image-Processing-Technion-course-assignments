```python
def PSNR(img1,psnr): 
    #creating the noise based on a given psnr value and adding the noise so that its PSNR will be psnr
    I1 = np.max(img1)
    sigma1 = I1 / (10**(psnr/20))
    noise = np.random.normal(I1,sigma1,img1.shape) #normal noise with mean of I_max and std calculate for the given psnr    
    #adding noise to the images
    img1_noisy = img1+noise 
    return img1_noisy , noise
```

2. Implement the `wiener_filter` function that recieves a blurry and noisy image, a filter kernel and a prior array and returns the restored image.<br>
The prior array is the believed signal-to-noise ratio of the Fourier transforms of the original image $(I)$ and noise $(N)$. The prior array will be what we believe is the ratio $\frac{N}{I}$.<br>
Don't forget to set the s parameter in the fft function.

```python
def wiener_filter(img: np.ndarray, kernel: np.ndarray, prior: np.ndarray) -> np.ndarray:
    '''
     The function recieves a blurry and noisy image, a filter kernel and a prior array and returns the restored image.
    
    :param img: blurry and noisy image
    :param kernel:a filter kernel
    :param prior: what we believe is the ratio  /
    :return img_restored:restored image
    '''
    # ====== YOUR CODE: ======
    G=np.fft.fft2(img)
    H=np.fft.fft2(kernel,s=img.shape)
    H_star=np.conjugate(H)
    F= (G*H_star) / (H_star*H + prior**2)
    img_restored = np.abs(fft.ifft2(F))
    # ======================== 
    return img_restored
```

In this exercise, you will Implement the Max-Lloyed algorithm for quantization of pixel values of an RGB image. 

Given a set of quantization levels $\left\{f_i^{(k)}\right\}$, we want to find a better set of quantization levels $\left\{f_i^{(k+1)}\right\}$. Finding and representing the borders of the quantization regions is hard in any dimension higher than 1D, therefore, instead of finding the decision levels, you will use the euclidean distance of each pixel value in the 3D RGB space for the quantization levels. A pixel value " belongs" to the quantization level closest to it. Next, find the new quantization levels by computing the mean pixel value belonging to each previous quantization level.      

1. Implement the function `max_lloyd_iter` that receives the image and a set of quantization levels and returns the next set of quantization levels. The quantization levels you are looking for are 3-dimensional vectors. Do not perform quantization for each color channel separately.  
Note: you can infer the dimension by assuming a 'channel last' convention. <br>
Write a description of your function and explain its inputs and output.   



```python
def max_lloyd_iter(img: np.ndarray, prev_levels: np.ndarray) -> np.ndarray:
    """
    Calculate the next set of quantization levels based on the given image and previous levels.
    
    :param img: The input image, represented as a 3-dimensional RGB numpy array.
    :param prev_levels: The previous set of quantization levels, represented as a 3-dimensional numpy array.
    :return: The next set of quantization levels.
    """
    # ====== YOUR CODE: ======
    # Get the number of rows and columns in the image
    num_rows, num_cols = img[:,:,0].shape
    
    # Initialize variables
    num_levels = prev_levels.shape[0]
    img_copy = img.copy().astype('int64')
    prev_levels_copy = prev_levels.copy().astype('int64')
    temp = np.zeros([num_levels,3])
    count = np.zeros(num_levels)
    
    # Compute the distance between each pixel and each level
    dists = np.linalg.norm(img_copy[:,:,None] - prev_levels_copy[None,None,:], axis=3)
    
    # Find the closest level to each pixel
    pix_levels = np.argmin(dists, axis=2)
    
    # Accumulate the sum and count for each level
    for i in range(num_levels):
        mask = (pix_levels == i)
        temp[i] = np.sum(img[mask], axis=0)
        count[i] = np.sum(mask)
    
    # Divide the sum by the count to get the new levels
    for j in range(num_levels):
        if count[j] == 0:
            count[j] = 1
    new_levels = temp/count[:,None]         
    return new_levels
```

To perform quantization, you have to initialize the quantization levels to some guess and iteratively improve that initial guess. You stop improving your quantization levels when a criterion you chose is met.<br>

2. Implement the function `max_lloyd_quantize` that receives an image and a number of quantization levels, performs Max-Lloyd quantization and returns the quantization levels and your metric values at each iteration.
    To ensure your algorithm does not run too long (in cases your threshold is too low), use the 'max_iter' parameter to stop your function prematurely. <br>
    Initialize the first level with a random value in the RGB space; for this initialization, set the 'init_type' parameter to 'random_rgb'.<br>
    The input and output data types have to be uint8.<br>
    Write a description of your function and explain its inputs and output. 


```python
#MSE between quatization levels at 2 succesive iterations

def max_lloyd_quantize(img: np.ndarray, level_num: int, threshold: float, max_iter: int,
                       init_type: str) -> Tuple[np.ndarray, np.ndarray]:
    '''
    receives an image and a number of quantization levels, performs Max-Lloyd quantization and
    returns the quantization levels and your metric values at each iteration.
    
    :param img: input omage
    :param level_num:number of quantization levels
    :param threshold:the quantization error threshold(MSE) 
    :param max_iter:max itertions allowed
    :param init_type: how to Initialize the quantization levels
    :return new_levels: new quantization levels
    :return metric_vals: matric values of all iterations
    '''
    # ====== YOUR CODE: ======
    # Initialize the quantization levels
    if init_type == 'random_rgb':
        levels = np.random.randint(0, 256, size=(level_num, 3)).astype(np.uint8)
    else:
        # Otherwise, initialize with random values from the image pixels (3D vectors)
        levels = init_levels(img,level_num).astype(np.uint8)
        
    new_levels = []
    metric_vals = []
    count_iter = 0
    MSE = threshold * 10
    while MSE > threshold and count_iter < max_iter:
        new_levels = max_lloyd_iter(img, levels)
        count_iter += 1
        MSE = np.mean((new_levels - levels) ** 2) 
        levels = new_levels
        metric_vals.append(MSE.astype('uint8'))
    # ========================
    return levels, metric_vals
```

```python
def img_quantize(img: np.ndarray, qunt_levels: np.ndarray,) -> np.ndarray: 
    '''
    Add your description and complete the inputs (params) and output (return).
    
    :param img: A RGB image - 3 dimentional matrix
    :param qunt_levels: Final quantization levels 
    :return qunt_img: The quantized version of the image with the quantization levels as values
    '''
    # ====== YOUR CODE: ======
    M, N = img[:, :, 0].shape
    qunt_img = np.zeros(img.shape, dtype=np.uint8)
    for row in range(M):
        for col in range(N):
            pix = img[row, col]
            best_level = None
            best_distance = float("inf")
            for level in qunt_levels:
                distance = np.linalg.norm(pix - level)
                if distance < best_distance:
                    best_distance = distance
                    best_level = level
            qunt_img[row, col] = best_level
    # ========================  
    return qunt_img
```

Implement the function `init_levels` that receives an image and a number of quantization levels (K) and returns a random choice of pixels vaules that can be used as initial quantization levels (a Kx3 array).<br>
Make sure you do not use the same pixel value twice. <br>
Write a description of your function and explain its inputs and output.<br>


```python
def init_levels(img: np.ndarray, level_num: int,) -> np.ndarray:
    '''
    receives an image and a number of quantization levels (K) and returns a random choice of pixels
    vaules that can be used as initial quantization levels (a Kx3 array).       
    :param img: A RGB image - 3 dimentional matrix
    :param level_num: Number of quantization levels
    :return init_vals: Iitial quantization levels - random values from the image
            
    '''
    # ====== YOUR CODE: ======
    M,N=img.shape[0],img.shape[1]
    
    img = img.reshape(-1, 3)    
    
    # Select K random indices
    indices = np.random.choice(np.arange(M*N), size=level_num, replace=False)
    
    # Extract the pixel values at the selected indices
    init_vals = img[indices]
    # ========================    
    return init_vals
```

**5. Why is it important to make sure you do not use the same quantization level twice? what will happen if you do use the same quantization level twice?**

**Answer:**
Using the same quantization level twice can lead to suboptimal quantization. If you use the same quantization level twice, it means that two different pixel values will be represented by the same quantization level. This can lead to a loss of detail in the image and an increase in the error between the original pixel values and the quantized pixel values. In the Max-Lloyd algorithm, the goal is to minimize the error between the original pixel values and the quantized pixel values, so using the same quantization level twice can lead to a higher error and a poorer quality quantized image.

6. Implement the function `img_quantize` that receives an image and quantization levels and creates a quantized version of the image.<br> 
Write a description of your function and explain its inputs and output.<br>


```python
def img_quantize(img: np.ndarray, qunt_levels: np.ndarray,) -> np.ndarray: 
    '''
    Add your description and complete the inputs (params) and output (return).
    
    :param img: A RGB image - 3 dimentional matrix
    :param qunt_levels: Final quantization levels 
    :return qunt_img: The quantized version of the image with the quantization levels as values
    '''
    # ====== YOUR CODE: ======
    M, N = img[:, :, 0].shape
    qunt_img = np.zeros(img.shape, dtype=np.uint8)
    for row in range(M):
        for col in range(N):
            pix = img[row, col]
            best_level = None
            best_distance = float("inf")
            for level in qunt_levels:
                distance = np.linalg.norm(pix - level)
                if distance < best_distance:
                    best_distance = distance
                    best_level = level
            qunt_img[row, col] = best_level
    # ========================  
    return qunt_img
```

Adding noise to an image before quantization can be beneficial in some cases because it can help reduce the loss of information that occurs during the quantization process.

Quantization involves reducing the number of unique colors in an image by mapping similar colors to a single representative color. This process can result in a loss of detail and introduce artifacts in the quantized image. By adding noise to the image before quantization, the colors in the image are perturbed, making them less similar to each other. As a result, the quantization process can better preserve the details in the original image because it has more colors to work with.

However, adding too much noise can also degrade the image quality, so it is important to find a balance between adding enough noise to preserve image details and not adding so much noise that it becomes distracting.
