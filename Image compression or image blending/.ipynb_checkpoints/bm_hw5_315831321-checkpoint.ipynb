1. Implement the function `symbol_prob` that receives a flattened image and returns a dictionary of symbols (key) and their probability (value).
Write a description of your function and explain its inputs and output.

```python
def symbol_prob(img: np.ndarray) -> dict:
    '''
    Add your description and complete the inputs (params) and output (return).
    This function create a dictionary with the image color values and their probability of appearance. 
    :param img: The flattened 2D Image 
    :return dict_probs: A dictionary with the image color value (key) and their probability (value). 
    '''
    # ====== YOUR CODE: ======
    dict_probs = dict()
    
    for value in img:
        if value in dict_probs:
            dict_probs[value] += 1
        else:
            dict_probs[value] = 1
    
    pixel_total = img.size
    
    for key in dict_probs:
        dict_probs[key] = dict_probs[key]/pixel_total
    # ========================     
    return dict_probs
```

```python
def bulid_tree(dict_prob: dict) -> list:
    '''
    Add your description and complete the inputs (params) and output (return).
    This function creat a representation of the Huffman Tree
    :param dict_prob: A dictionary with the image color value (key) and their probability (value). 
    :return tree: The huffman tree
    '''
    # ====== YOUR CODE: ======
    def takeSecond(elem):
        return elem[1]

    length = len(dict_prob)
    for i in range(length-1):
        
        if i == 0:
            
            dict_sort = sorted(dict_prob.items(), key=lambda x: x[1])
        if i != 0:
            
            dict_sort = save
            dict_sort.sort(key=takeSecond)
            
        dict_sort.append((list((dict_sort[0][0], dict_sort[1][0])), (dict_sort[0][1]+dict_sort[1][1])))
        dict_sort = dict_sort[2:]
        save = dict_sort
        
    tree = list(save[0][0])
    # ========================  
    return tree
```

3. Implement the function `huffman code` that receives the huffman tree and returns a dictionary of the symbols and their huffman coding (a binary string). <br> If needed, you may add an additional input to the function. 

    Optional guide: Assign an empty string to the root and then recursively assign 1 for the right branch and  0 for the left branch.<br>

```python
def huffman_code(huffman_tree: list, String: str) -> dict:
    '''
    Add your description and complete the inputs (params) and output (return).
    This function implement the binary string for each of the color values.
    :param huffman_tree: The huffman tree
    :return dict_code: A dictionary with the image color value (key) and their binary code (value).
    '''
    # ====== YOUR CODE: ======
    dict_code = {}
    if type(huffman_tree) != list:
        dict_code[huffman_tree] = String
    else:
        dict_code.update(huffman_code(huffman_tree[0], String+"0"))
        dict_code.update(huffman_code(huffman_tree[1], String+"1"))
    # ========================         
    return dict_code
```

4. Implement the function `huffman_encode` that receives the huffman code and the flatten image and returns the encoded image.

```python
def huffman_encode(code: dict, img: np.ndarray) -> list: 
    '''
    Add your description and complete the inputs (params) and output (return).
    This function receives the huffman code and the flatten image and returns the encoded image
    :param code: A dictionary with the image color value (key) and their binary code (value).
    :params img: The flattened 2D Image 
    :return encoded_img: The encoded image
    '''
    # ====== YOUR CODE: ======
    encoded_img = ['']*len(img)
    
    for key, value in enumerate(img):
        encoded_img[key] = code[value]
    # ========================     
    return encoded_img
```

```python
def huffman_decode(encoded_img:list, tree: list) -> np.array: 
    '''
    Add your description and complete the inputs (params) and output (return).
    This function receives the encoded image and the huffman tree of the image and returns the decoded image.
    
    :param encoded_img: The encoded image
    :params tree: A dictionary with the image color value (key) and their binary code (value).
    :return decoded_img: The decoded image
    '''
    # ====== YOUR CODE: ======
    length = len(encoded_img)
    decoded_img = np.zeros((length,1))
    for i, j in enumerate(encoded_img):
        for key, value in tree.items():
             if j == value:
                    decoded_img[i] = key
    # ========================
    return decoded_img
```

1. First, implement the `gaussian_pyramid`  function that creates a list of downscaled versions of the input image by a scale of 2. Before each down sampling, use gaussian smoothing to avoid aliasing.

    Note: The smoothing is required to avoid aliasing. Do not smooth the mask so that it will remain a binary mask.

    Do not use the resize functions for downscaling. You may use the 'gaussian' function from skimage.filters
    
    Write a description of your function and explain its inputs and output.

```python
def gaussian_pyramid(
        img: np.ndarray,
        num_levels: int,
        mask: bool = False,
        sigma: float = 0.4,
    ) -> list:
    '''
    Add your description and complete the inputs (params) and output (return).
    This function create Gaussian Pyramid. following images are weighted down and scaled down using a Gaussian average (Gaussian blur). 
    Every pixel with a local average corresponds to a neighbor pixel on a lower level of the pyramid.
    
    :param img: RGB image
    :params num_levels: Number of layers of the pyramid
    :params mask: boolean requirement, if the img parameter is a mask do not use gaussian filter
    :params sigma: the SD of the gaussian filter distribution
    :return gaussian_pyr: list of downscaled versions of the input image by a scale of 2.
    '''
    # ====== YOUR CODE: ======
    save = (img/255).astype('float64')
    gaussian_pyr = [save]

    if mask:
        for i in range(num_levels):
            down_img = save[0::2,0::2,:]
            save = down_img
            gaussian_pyr.append(down_img)
    else:    
        for i in range(num_levels):
            filtered_image = np.zeros(save.shape)
            filtered_image[:,:,0]  = gaussian(save[:,:,0], sigma = sigma)
            filtered_image[:,:,1]  = gaussian(save[:,:,1], sigma = sigma)
            filtered_image[:,:,2]  = gaussian(save[:,:,2], sigma = sigma)
            filtered_image = filtered_image[0::2,0::2,:]
            save = (filtered_image).astype('float64')
            gaussian_pyr.append(save)
    # ========================
    return gaussian_pyr
        
```

2. Implement the `laplacian_pyramid` function that receives a Gaussian pyramid and computes an approximation of the Laplacian using the difference of Gaussians in every two consecutive levels of the pyramid. The top of the pyramid (lowest resolution) will remain the same.

    You may use the 'resize' function from skimage.transform

    Write a description of your function and explain its inputs and output.
    
    **Bonus (5 points):**   
Instead of using the 'resize' function from skimage, use your image transformation functions from homework assignments 3. Only instead of rotation, perform scaling.

```python
def back_project(size: tuple) -> np.ndarray:
    '''
    This function produce coordinates of the corresponding donor pixel in the source image
    
    :param mapping: 3x3 transformation matrix
    :param size: The destination image size
    :return source_coords: The coordinates of the transformed destination pixels in the source image

    '''
    # ====== YOUR CODE: ======
    # create the scaling transformation matrix
    mapping = np.array([[2, 0, 0], [0, 2, 0], [0, 0, 1]]) 
    
    # create array of destination x,y coordinate
    h ,w = size
    X,Y = np.meshgrid(np.linspace(0, w-1,w), np.linspace(0,h-1,h))
    source_coords = np.zeros([h , w,2])
    
    # calculate source coordinates
    invT = np.linalg.inv(mapping)
    coords = np.column_stack((Y.flatten(), X.flatten(), np.ones(Y.shape).flatten())).T # 3xN
    mat = np.dot(invT,coords)[0:2,:]
    source_coords[:,:,0] = np.reshape(mat[0,:],(h ,w))
    source_coords[:,:,1] = np.reshape(mat[1,:],(h ,w))
    # ========================
    return source_coords
```

```python
def bilinear_interpolation(source: np.ndarray, coords: np.ndarray) -> np.ndarray:
    '''
    Add your description and complete the inputs (params) and output (return).
    
    :param source: The source image
    :param coords: The coordinates of the transformed destination pixels in the source image
    :return interpolation: The target image after implement the transformation
    '''
    # ====== YOUR CODE: ======
    interpolation = np.zeros(coords[:,:,0].shape)
    M,N = source.shape
    h,w,c = coords.shape
    for col in range(h):
        for row in range(w):
            y, x = coords[col, row, 0], coords[col, row, 1]
            if 0<y<M-1 and 0<x<N-1:
                NW = (int(np.floor(y)),int(np.floor(x)))
                NE = (int(np.floor(y)),int(np.ceil(x)))
                SW = (int(np.ceil(y)),int(np.floor(x)))
                SE = (int(np.ceil(y)),int(np.ceil(x)))
                u = x - int(np.floor(x))
                v = int(np.ceil(y))-y
                S_u = source[SE]*u + source[SW]*(1-u)
                N_u = source[NE]*u + source[NW]*(1-u)
                interpolation[col,row] += N_u*v + S_u*(1-v)               
    # ========================        
    return interpolation
```

```python
def laplacian_pyramid(gaussian_pyr: list) -> list:
    '''
    Add your description and complete the inputs (params) and output (return).
    This function saves the difference image between each level's blurred versions. 
    Only the lowest level lacks a difference image, 
    allowing the high resolution image to be rebuilt using the difference photos on higher levels.
    
    :param gaussian_pyr: list of downscaled versions of the input image by a scale of 2.
    :return laplacian_pyr: list of different matrices representing different frequencies of the input image.
    '''
    # ====== YOUR CODE: ======
    laplacian_pyr = []
    for i in range(len(gaussian_pyr)-1):
        save = np.zeros(gaussian_pyr[i].shape)
        save = resize(gaussian_pyr[i+1], gaussian_pyr[i].shape)
        laplacian_pyr.append((gaussian_pyr[i]-save))
    laplacian_pyr.append(gaussian_pyr[len(gaussian_pyr)-1])   
    # ======================== 
    return laplacian_pyr
```

```python
def laplacian_pyramid_w_HW3(gaussian_pyr: list) -> list:
    '''
    Add your description and complete the inputs (params) and output (return).
    This function saves the difference image between each level's blurred versions. 
    Only the lowest level lacks a difference image, 
    allowing the high resolution image to be rebuilt using the difference photos on higher levels.
    
    :param gaussian_pyr: list of downscaled versions of the input image by a scale of 2.
    :return laplacian_pyr: list of different matrices representing different frequencies of the input image.
    '''
    # ====== YOUR CODE: ======
    laplacian_pyr = []
    for i in range(len(gaussian_pyr)-1):
        save = np.zeros(gaussian_pyr[i].shape)
        coords_r = back_project(gaussian_pyr[i][:,:,0].shape)
        coords_g = back_project(gaussian_pyr[i][:,:,1].shape)
        coords_b = back_project(gaussian_pyr[i][:,:,2].shape)
        save[:,:,0] = bilinear_interpolation(gaussian_pyr[i+1][:,:,0], coords_r)
        save[:,:,1] = bilinear_interpolation(gaussian_pyr[i+1][:,:,1], coords_g)
        save[:,:,2] = bilinear_interpolation(gaussian_pyr[i+1][:,:,2], coords_b)
        laplacian_pyr.append((gaussian_pyr[i]-save))
    laplacian_pyr.append(gaussian_pyr[len(gaussian_pyr)-1])   
    # ======================== 
    return laplacian_pyr
```

3. Implement the `blend` function that receives 3 Laplacian pyramids, those of the two images you want to stitch together and that of the mask, and blend the two pyramids according to the mask. At each level of resolution, the combined pyramid will contain the pixels in the mask (ones) from the first pyramid and the pixels from out of the mask (zeros) from the second pyramid.

    Write a description of your function and explain its inputs and output.

```python
def blend(laplacian_B: list, laplacian_A: list, mask_pyr: list) -> list:
    '''
    Add your description and complete the inputs (params) and output (return).
    
    :param laplacian_A: list of different matrices representing different frequencies of the input image
    :params laplacian_B: list of different matrices representing different frequencies of the input image
    :params mask_pyr: list of different matrices representing different frequencies of the input image
    :return blended_imgs: the combined pyramid will contain the pixels in the mask (ones) 
            from the first pyramid and the pixels from out of the mask (zeros) from the second pyramid
    '''
    # ====== YOUR CODE: ======
    blended_imgs = []

    for i in range(len(laplacian_A)):
        save = laplacian_A[i] - np.multiply(laplacian_A[i],mask_pyr[i])
        a = save + np.multiply(laplacian_B[i],mask_pyr[i])
        blended_imgs.append(a)
    # ========================  
    return blended_imgs
```

4. Implement the `reconstruct` function that receives a Laplacian pyramid and reconstructs the Gaussian pyramid that was used to construct it by upscaling the top of the pyramid (lowest resolution) and adding it to the next layer and doing the same with the first reconstructed level to reconstruct the next layer and so on for all layers.

     You may use the 'resize' function from skimage.transform
     
     Write a description of your function and explain its inputs and output.
  

```python
def reconstruct(laplacian_pyr: list,) -> list:
    '''
    Add your description and complete the inputs (params) and output (return).
    This function reconstruct the blending image between the two images.
    :param laplacian_pyr: the combined pyramid will contain the pixels in the mask (ones) 
            from the first pyramid and the pixels from out of the mask (zeros) from the second pyramid
    :return recon_pyr: The reconstructed gaussian pyramid
    '''
    # ====== YOUR CODE: ======
    recon_pyr = [laplacian_pyr[-1]]
    
    for i in range(len(laplacian_pyr)-1):
        resized = resize(recon_pyr[i], np.shape(laplacian_pyr[-(i+2)]))
        a = (resized+laplacian_pyr[-(i+2)]) - np.min((resized+laplacian_pyr[-(i+2)]))
        a = a/np.max(a)
        recon_pyr.append(a)
    # ========================
    return recon_pyr        
```
